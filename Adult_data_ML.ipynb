```python
ip=dt.drop('income',axis=1)
op = dt.income
```

train testsplit

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,train_size=0.8)
```

standard scalar

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# machine learning model : logistic regression 
it simple trying to convert the independent variable to a expression of probability with the range of0s and 1s with respect to the dependent variable.
sigmoid function :
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

Prediction

```python
pred = lr.predict(x_test)
pred
```

Accuracy 

confusion matrics

a confusion matrics is a table used to evaluate the performance of a classifaction model it shows you how well your model predicted the actual classes by comparing :
- what the mmodel predicted
- what the actual correct answer were.

Accuracy tells u how many total prediction wait correct
- formulaa:   correct prediction / total prediction
       TN +TP/TN+FP+FN+TP

f1-score :
-it is the harmonic mean of precision and recall 
-2*(recll*precision/ recall + precision)

-good when u need a balanced between precision and recall .
-it balanced both FP and FN

Precision:- it tells u how many predictive positives were  actually correct .   prediction Actual Positive/total predicted positive

recall:- it tells u how many actual positive were correctly found.
    - prediction Actual positive/total actual positive

#accuracy....
Accuracy : overallcorectness
precision : correction of positives
recall: how many actuall positive are cought
f1_score :balancee between precision and recall.

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
acc = accuracy_score(pred,y_test)
recall = recall_score(pred,y_test)
f1 = f1_score(pred,y_test)
precision= precision_score(pred,y_test)
print(f'Accuracy score :{acc}')
print(f'Recall score :{recall}')
print(f'f1 score :{f1}')
print(f'Precision score :{precision}')
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(pred,y_test)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
```

```python
cmd.plot()
```
